```python
import sqlite3
import pandas as pd

DB_PATH = "../data/database.db"

def run_query(query):
    conn = sqlite3.connect(DB_PATH)
    df = pd.read_sql(query, conn)
    conn.close()
    return df

top_rated_movies = run_query("""
    SELECT title, year, vote_average 
    FROM movies 
    WHERE vote_count >= 50 
    ORDER BY vote_average DESC 
    LIMIT 10
""")

print(top_rated_movies)
```
